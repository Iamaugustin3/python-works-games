```python
def player_input():
    
    f=0
    while f==0:
        marker=input("Please enter the marker of Player 1('X'or'O'):")
        if marker not in ('X','O'):
            print('Enter a valid marker..!!')
        elif marker=='X':
            player1='X'
            player2='O'
            f=1
        else:
            player1='O'
            player2='X'
            f=1
    return (player1,player2)
```

```python
def board_position():
    choice='k'
    num=range(0,10)
    while choice.isdigit()==False:
        
        choice=input("Choose a position(1-9): ")
        if choice.isdigit()==False:
            print("Please enter a digit")
        elif int(choice) not in num:
            print("Kindly enter a value between 1 and 9..!!")
            choice='k'
    return(choice)
```

```python
def place_marker(board, marker, position):
    
    #placing marker on the board
    marker=input("Enter the marker('X'or'O':)")
    if marker not in ('X','O'):
        print('Enter a valid marker..!')
    else:
        board[position]=marker
```

```python
def win_check(board, mark):
    
    return ((board[1]==mark and board[2]==mark and board[3]==mark) or
            (board[4]==mark and board[5]==mark and board[6]==mark) or
            (board[7]==mark and board[8]==mark and board[9]==mark) or
            (board[1]==mark and board[5]==mark and board[9]==mark) or
            (board[3]==mark and board[5]==mark and board[7]==mark) or
            (board[1]==mark and board[4]==mark and board[7]==mark) or
            (board[3]==mark and board[6]==mark and board[9]==mark))

    
```

```python
import random

def choose_first():
    return(random.randint(1,2))
```

```python
def space_check(board, position):
    
    return(board[position]==' ')
```

```python
def full_board_check(board):
    # To check the whole board is full, to check if its a draw
    for i in range(1,10):
        if space_check(board,i):
            return False
    return True
```

```python
def player_choice(board):
    
    choice='k'
    num=range(0,10)
    while choice.isdigit()==False or space_check(board,int(choice))==False:
        
        choice=input("Choose a position(1-9): ")
        if choice.isdigit()==False:
            print("Please enter a digit")
        elif int(choice) not in num:
            print("Kindly enter a value between 1 and 9..!!")
            choice='k'
        elif space_check(board,int(choice))==False:
            choice=input("Not a free position, Choose another position (1-9):")
    return int(choice)
```

```python
def replay():
    k=0
    while k==0:
        c=input("Do you want to play again?(Y/N) :")
        if c=='Y':
            return True
            k=1
        elif c=='N':
            return False
            k=1
        else:
            print("Enter a valid answer")
```

```python
print('Welcome to Tic Tac Toe!')

while True:
#     Settin' the game up 
    the_board=[' ']*10
    display_board(the_board)
    player1_marker,player2_marker=player_input()
    turn=choose_first()
    print('Player {} will go first'.format(turn))
    play_game=input("Ready to play the game? (y/n): ")
    if play_game=='y':
        game_on=True
    else:
        game_on=False
    while game_on:
        clear_output
        
        if turn==1:
            display_board(the_board)
            
#         Player 1 Turn

            position=player_choice(the_board)
            place_marker(the_board,player1_marker,position)
            if win_check(the_board,player1_marker)==True:
                display_board(the_board)
                print('Player 1 is the winner')
                game_on=False
            else:
                if full_board_check(the_board)==True:
                    print('It is a tie')
                    game_on=False
                else:
                    turn=2
#         Player 2 Turn
    
        else:
            display_board(the_board)
            position=player_choice(the_board)
            place_marker(the_board,player2_marker,position)
            if win_check(the_board,player2_marker)==True:
                display_board(the_board)
                print('Player 2 is the winner')
                game_on=False
            else:
                if full_board_check(the_board)==True:
                    print('It is a tie')
                    game_on=False
                else:
                    turn=1
            

    if not replay():
        break
```
