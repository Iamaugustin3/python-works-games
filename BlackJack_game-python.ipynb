```python
import random
global playing
playing=True
suits = ('Hearts','Diamonds','Spades','Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,
          'King':10,'Ace':11}
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
    
    def __str__(self):
        return self.rank+' of '+self.suit
```

```python
test_card=Card('Spades','Ace') #testing
print(test_card)
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []  # to start with an empty list
        for suit in suits:
            for rank in ranks:
                
                #appending all cards to form a deck 
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        deck_all=''
        for card in self.deck:
            deck_all+= '\n'+card.__str__()
        return 'The deck has: '+deck_all
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        new_card=self.deck.pop()
        return new_card
```

```python
class Hand:
    def __init__(self):
        self.cards = []  # to start with an empty list ~ I did in the Deck class
        self.value = 0   # starting with a zero value
        self.aces = 0    # added an attribute to keep track of aces
    
    def add_card(self,card):
        self.cards.append(card)
        self.value+=values[card.rank]
        
        if self.cards=='Ace':
            self.aces+=1
        
    def adjust_for_ace(self):
        
        while self.value>21 and self.aces:
            self.value-=10
            self.aces-=1
```

```python
class Chips:
    
    def __init__(self,total=100):
        self.total = total  # This can be set to a default value or user can assign a value
        self.bet = 0
        
    def win_bet(self):
        self.total+=self.bet
    
    def lose_bet(self):
        self.total-=self.bet
```

```python
def take_bet(chips):

    while True:
    
    #Using try and except
        try:
            chips.bet=int(input('Enter the number of chips you want to bet: '))
        except:
            print('Sorry, wrong input..! must be an integer')
        else:
            if chips.bet>chips.total:
                print('Sorry, you dont have enough chips, you have: {}'.format(chips.total))
            else:
                break
```

```python
def hit(deck,hand):
    #function calls on hit
    one_card=deck.deal()
    hand.add_card(one_card)
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    
    global playing  # to control an upcoming while loop
    
    while True:
        
        k=input('\nEnter your choice, Is it a hit or a stand (h/s): \n')
        if k[0].lower()=='h':
            hit(deck,hand)
            
        elif k[0].lower()=='s':
            print('Player chose to stand, Dealers turn..!\n')
            playing=False
                  
        else:
            print('Please enter a valid choice..!')
            continue
                  
        break
```

```python
def show_some(player,dealer):
    
    print("\nDealers Hand:\n")
    print('One card is hidden.!\n')
    print(dealer.cards[1])
    print('\n')
    print('\nPlayers Hand:\n')
    for card in player.cards:
        print(card)
                            #func to show player's and dealer's hand
    
    
def show_all(player,dealer):
    
    print('\nDealers Hand:\n')
    for card in dealer.cards:
        print(card)
    print('\n')
    print('\nPlayers Hand:\n')
    for pcard in player.cards:
        print(pcard)
```

```python
def player_busts(player,dealer,chips):
    print('Dealer wins, Player Busted')
    chips.lose_bet()
    
def player_wins(player,dealer,chips):
    print('Player Wins..!')
    chips.win_bet()
    
def dealer_busts(player,dealer,chips):
    print('Player Wins, Dealer Busted..!')
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print('Dealer Wins..!')
    chips.lose_bet()
    
def push(player,dealer):
    print('Its a tie..!, Nobody Wins')
```

```python
while True:
  
    print('Welcome to BLACKJACK')
    
    # Creating & shuffling the deck, dealing two cards to each player
    
    deck=Deck()
    deck.shuffle()
    player_hand=Hand()
    player_hand.add_card(deck.deal())
    player_hand.add_card(deck.deal())
    
    dealer_hand=Hand()
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    
        
    # Setting up the Player's chips
    
    player_chips=Chips()
    
    
    # Prompting the Player for their bet
    take_bet(player_chips)
    
    # Showing player and dealers cards (one dealer card is hidden)
    show_some(player_hand,dealer_hand)
    
    
    while playing:  # variable from the hit_or_stand function
        
        # Prompting for Player to Hit or Stand
        hit_or_stand(deck,player_hand)
        
        # Showing player and dealers cards (one dealer card is hidden)
        show_some(player_hand,dealer_hand)
        
        # If player's hand exceeds 21, calling player_busts() and breaking out of loop
        if player_hand.value>21:
            player_busts(player_hand,dealer_hand,player_chips)
            break

    # If Player hasn't busted, playing Dealer's hand until Dealer reaches 17
    if player_hand.value<=21:
        
        while dealer_hand.value<17:
            hit(deck,dealer_hand)
    
        # Showing all cards( both dealer and player)
        show_all(player_hand,dealer_hand)
        # showing different winning scenarios
        if dealer_hand.value>21:
            dealer_busts(player_hand,dealer_hand,player_chips)
        elif dealer_hand.value>player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chips)
        elif player_hand.value>dealer_hand.value:
            player_wins(player_hand,dealer_hand,player_chips)
        else:
            push(player_hand,dealer_hand)
    
    # Informing Player about total chips in his/her hand
    print('\nYou have a total of {} chips'.format(player_chips.total))

    n=input('Would you like to play again?.!(y/n)') #to play again
    if n[0].lower()=='y':
        playing=True
        continue
    else:
        print('Thanks for playing..!')
        break
```
